# Data cleaning and analysis

```python
# Importing python libraries and assigning the data to a variable
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
df = pd.read_csv("TWO_CENTURIES_OF_UM_RACES_1.csv")
df.head(20)
```

```python
# Getting data info
df.info()
```

```python
# getting the bottom part of the data
df.tail(20)
```

```python
# Assigning the data set to a variable while locating all  data assigned to female under the Athlet gender column
df2 = df.loc[df['Athlete gender'] == 'F']
df2
```

```python
# Getting the new dataset information
df2.info()
```

```python
# Assigning the dataset to a new variable while getting data associated with 'km' and 'mi' in Event distance/length
df3 = df2[df2["Event distance/length"].str.contains("km")| df2["Event distance/length"].str.contains("mi")]
df3
```

```python
# Creating new columns for distances in 'km' and 'mi'
df3['Event distance/length in km'] = df.apply(lambda _: None, axis = 1)
df3['Event distance/length in mi'] = df.apply(lambda _: None, axis = 1)
df3
```

```python
# Data information
df3.info()
```

```python
# Split 'Event distance/length' column by 'k' and 'm' to get the values into separate columns
for i in range(len(df3['Event distance/length'])):
    if 'km' in df3['Event distance/length'].values[i]:
        arr1 = df3['Event distance/length'].values[i].split('k')
        df3['Event distance/length in km'].values[i] = arr1[0]
    elif 'mi' in df3['Event distance/length'].values[i]:
        arr2 = df3['Event distance/length'].values[i].split('m')
        df3['Event distance/length in mi'].values[i] = arr2[0]
df3
```

```python
# Creating a new column named 'Event country' while putting the respective values using the split method
df3['Event country'] = df.apply(lambda _: None, axis = 1)
for i in range(len(df3['Event name'])):
    if '(' in df3['Event name'].values[i]:
        arr1 = df3['Event name'].values[i].split('(')
        df3['Event name'].values[i] = arr1[0]
        df3['Event country'].values[i] = arr1[1]
df3
```

```python
# Finishing up the values addition to the new column
for i in range(len(df3['Event country'])):
    if ')' in df3['Event country'].values[i]:
        arr1 = df3['Event country'].values[i].split(')')
        df3['Event country'].values[i] = arr1[0]
df3
```

```python
# Data information
df3.info()
```

```python
# Clean the data
df3['Event distance/length in mi'] = df3['Event distance/length in mi'].str.replace(",", ".")
```

```python
# Changing every 'None' values to zero
for i in range(len(df3['Event distance/length in mi'])):
    if df3['Event distance/length in mi'].values[i] is None:
        df3['Event distance/length in mi'].values[i] = '0'

df3
```

```python
# Changing 'Event distance/length in mi' column to float datatype
df3['Event distance/length in mi'] = df3['Event distance/length in mi'].replace('[^\d.]', '', regex = True).astype(float)
```

```python
# Data information
df3.info()
```

```python
# Replacing a string with another string in 'Event distance/length in km' column 
df3['Event distance/length in km'] = df3['Event distance/length in km'].str.replace(",", ".")
```

```python
# Getting unique values of 'Event distance/length in km' column
df3['Event distance/length in km'].unique()
```

```python
# Converting values in miles to kilometers
a_mile = 1.609
df3['Event distance/length in mi'] = df3['Event distance/length in mi'] * a_mile
df3
```

```python
# Changing 'None' values in 'Event distance/length in km' to zero
for i in range(len(df3['Event distance/length in km'])):
    if df3['Event distance/length in km'].values[i] is None:
        df3['Event distance/length in km'].values[i] = '0'

df3
```

```python
# Changing the datatype of 'Event distance/length in km' to float
df3['Event distance/length in km'] = df3['Event distance/length in km'].replace('[^\d.]', '', regex = True).astype(float)
```

```python
# Dropping null values in 'Athlete year of birth' column
df3.dropna(subset = 'Athlete year of birth', inplace = True)
df3
```

```python
# Changing datatypes of some columns to integer
df3['Athlete year of birth'] = df3['Athlete year of birth'].astype(int)
df3['Year of event'] = df3['Year of event'].astype(int)
```

```python
# Creatng a new column 'Athlete age' and getting its required values
df3['Athlete age'] = df.apply(lambda _: None, axis = 1)
df3['Athlete age'] = df3['Year of event'] - df3['Athlete year of birth']
df3
```

```python
# Getting the values of 'Athlete age category' 
df3['Athlete age category'].unique()
```

```python
# Dropping null cells in 'Athlete age category' column
df3.dropna(subset = 'Athlete age category', inplace = True)
```

```python
# Stripping 'Athlete age category' column of unwanted strings
df3['Athlete age category'] = df3['Athlete age category'].str.lstrip('W')
df3['Athlete age category'] = df3['Athlete age category'].str.lstrip('U')
df3['Athlete age category'] = df3['Athlete age category'].str.lstrip('F')
df3
```

```python
# Changing 'Athlete age category' column to integer
df3['Athlete age category'] = df3['Athlete age category'].apply(int)
df3.info()
```

```python
# Creating lists and appending values into them

Eligibility1 = []
Eligibility2 = []
for i in range(len(df3)):
    if df3['Athlete age'].values[i] <= df3['Athlete age category'].values[i]:
        Eligibility1.append('Eligible')
    else:
        Eligibility2.append('Not eligible')
    
    
print(len(Eligibility1))
print(len(Eligibility2))
```

```python
# Creating a new column and assigning the necessary values
Eligible = []
df3['Eligible'] = df.apply(lambda _: None, axis = 1)
for i in range(len(df3)):
    if df3['Athlete age'].values[i] <= df3['Athlete age category'].values[i]:
        Eligible.append('Eligible')
    else:
        Eligible.append('Not eligible')
        
df3['Eligible'] = Eligible
```

```python
# Getting the final 'Event distance/length' values in km
df3['Event distance/length'] = df3['Event distance/length in km'] + df3['Event distance/length in mi']
df3
```

```python
# Stripping 'New athlete performance' column of the string 'h'
df3['New athlete performance'] = df3['Athlete performance'].str.rstrip('h')
df3
```

```python
# Creating new columns and assigning the necessary values to them
df3['Time in hours'] = df.apply(lambda _: None, axis = 1)
df3['Time in minutes'] = df.apply(lambda _: None, axis = 1)
df3['Time in seconds'] = df.apply(lambda _: None, axis = 1)

for i in range(len(df3['New athlete performance'])):
    if ':' in df3['New athlete performance'].values[i]:
        arr1 = df3['New athlete performance'].values[i].split(':')
        df3['Time in hours'].values[i] = arr1[0]
        df3['Time in minutes'].values[i] = arr1[1]
        df3['Time in seconds'].values[i] = arr1[2]
df3
```

```python
# Creating a new column and assigning the necessary values to it
df3['Time in days'] = df.apply(lambda _: None, axis = 1)

for i in range(len(df3['Time in hours'])):
    if 'd' in df3['Time in hours'].values[i]:
        arr1 = df3['Time in hours'].values[i].split('d')
        df3['Time in days'].values[i] = arr1[0]
        df3['Time in hours'].values[i] = arr1[1]
df3
```

```python
# Creating an empty list and assigning values to the list
hrs = []
for i in df3['Time in hours'].unique():
    hrs.append(i)
hrs   
```

```python
# Getting the unique values of 'Time in days'
df3['Time in days'].unique()
```

```python
# Splitting 'Time in seconds' column by 'k' and assigning values to it
for i in range(len(df3['Time in seconds'])):
    if 'k' in df3['Time in seconds'].values[i]:
        arr1 = df3['Time in seconds'].values[i].split('k')
        df3['Time in seconds'].values[i] = arr1[0]
df3
```

```python
# Getting the unique values
df3['Time in seconds'].unique()
```

```python
# Changing datatypes of the following columns to float
df3['Time in days'] = df3['Time in days'].astype(float)
df3['Time in hours'] = df3['Time in hours'].astype(float)
df3['Time in minutes'] = df3['Time in minutes'].astype(float)
df3['Time in seconds'] = df3['Time in seconds'].astype(float)
```

```python
# Creating a list, appending values and getting the length of the list
r = []
for i in df3['Time in hours'].unique():
    r.append(i)
len(r)
```

```python
# Getting the unique values of 'Time in minutes' column
df3['Time in minutes'].unique()
```

```python
# Creating a new column 'days to hrs' and changing 'None' values in 'Time in days' to zero
df3['days to hrs'] = df.apply(lambda _: None, axis = 1)
for i in range(len(df3['Time in days'])):
    if df3['Time in days'].values[i] == None:
        df3['Time in days'].values[i] == 0

df3
```

```python
# Replacing null values in 'Time in days' column with '0'
df3['Time in days'] = df3['Time in days'].fillna(0)

df3
```

```python
# Converting 'days to hrs', 'Time in seconds' and 'Time in minutes' columns to time in hours
df3['days to hrs'] = df3['Time in days'] * 24
df3['Time in seconds'].round(1) == df3['Time in seconds'] * 0.000278
df3['Time in minutes'].round(1) == df3['Time in minutes'] * 0.015873
df3
```

```python
# Getting unique values
df3['days to hrs'].unique()
```

```python
# Getting the final athlete performance
df3['Final athlete performance'] = df3['Time in hours'] + df3['Time in minutes'] + df3['Time in seconds'] + df3['days to hrs']
df3
```

```python
# Getting unique values
df3['Final athlete performance'].unique()
```

```python
# Creating a new column 'Athlete speed' and getting the required values
df3['Athlete speed'] = df.apply(lambda _: None, axis = 1)
df3['Athlete speed'] = df3['Event distance/length'] / df3['Final athlete performance']
df3['Athlete speed'] = df3['Athlete speed'].round(1)
df3
```

```python
# Data information
df3.info()
```

```python
# Creating a list and appending values into the list
date = []
for i in df3['Event dates'].unique():
    date.append(i)
date
```

```python
mean_speed = df3.groupby(['Athlete country'])['Athlete speed'].mean().reset_index()

# Group by 'Athlete country' and calculate the count of appearances
appearances = df3.groupby(['Athlete country']).size().reset_index(name='Appearances')

# Merge the two Dataframes on 'Athlete country'
App_spd = pd.merge(mean_speed, appearances, on='Athlete country')

# Print the result
App_spd
```

```python
# Normalize average speed and appearances
App_spd['Normalized Speed'] = (App_spd['Athlete speed'] - App_spd['Athlete speed'].min()) / (App_spd['Athlete speed'].max() - App_spd['Athlete speed'].min())
App_spd['Normalized Appearances'] = (App_spd['Appearances'] - App_spd['Appearances'].min()) / (App_spd['Appearances'].max() - App_spd['Appearances'].min())

# Define weight factors
weight_speed = 0.5
weight_appearances = 0.5

# Calculate weighted score
App_spd['Weighted Score'] = weight_speed * App_spd['Normalized Speed'] + weight_appearances + App_spd['Normalized Appearances']

# Calculate performance score
App_spd['Performance Score'] = App_spd['Weighted Score']

# Sort by performance score
App_spd = App_spd.sort_values(by='Performance Score', ascending=False)

# Print the result
App_spd
```

```python
df9 = App_spd.groupby('Athlete country')[['Weighted Score']].mean()
top_countries = df9.nlargest(10, 'Weighted Score')
top_countries
```

```python
# Visualization of the best countries
top_countries.plot(kind = 'bar', color = 'orange')
plt.ylabel('Performance Score')
plt.xlabel('Athlete country')
plt.title('Countries with the best performances')
plt.show
```
