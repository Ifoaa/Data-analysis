# Data Cleaning and Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('owid-covid-data.csv')
df
```

```python
df.describe()
```

```python
df.info()
```

```python
df.drop_duplicates(keep = False, inplace = True)
df
```

```python
df.dropna(subset='total_deaths', inplace = True)
df
```

```python
df.drop(['icu_patients', 'icu_patients_per_million', 'hosp_patients', 'hosp_patients_per_million', 'weekly_icu_admissions', 'weekly_icu_admissions_per_million', 'weekly_hosp_admissions', 'weekly_hosp_admissions_per_million', 'total_tests'], inplace = True, axis = 1)
```

```python
df.info()
```

```python
df.dropna(subset = 'reproduction_rate', inplace = True)
```

```python
df.info()
```

```python
df.dropna(subset = 'continent', inplace = True)
```

```python
df.info()
```

```python
df.dropna(subset = 'new_cases_smoothed', inplace = True)
```

```python
df.info()
```

# Data Visualization

```python
# Continent with the highest number of deaths
df1 = df.groupby(['location', 'continent']) ['total_deaths'].count().to_frame().sort_values(by = 'total_deaths', ascending = True)
df1
```

```python
df1.head(20).plot(kind = 'bar', color = 'green')
plt.xlabel('Continent', size = 5)
plt.ylabel('Total Deaths')
plt.title('Continent with the highest number of Deaths')
```

```python
df2 = df.groupby(['continent', 'location']) ['new_cases'].count().to_frame().sort_values(by = 'new_cases', ascending = True)
df2
```

```python
df2.head(20).plot(kind = 'bar', color = 'red')
plt.xlabel('Continent', size = 5)
plt.ylabel('New Cases', size = 5)
plt.title("Continent with the highest number of New cases")
plt.show()
```

```python
df3 = df.groupby(['continent', 'location']) ['new_cases'].sum().to_frame().sort_values(by = 'new_cases', ascending = False)
df3
```

```python
df3.head(20).plot(kind = 'bar', color='purple')
plt.xlabel('Location', size = 5)
plt.ylabel('New Cases', size = 5)
plt.title("Continent with the highest New Cases")
plt.show()
```

```python
df4 = df.groupby(['continent', 'location']) ['total_deaths'].sum().to_frame().sort_values(by = 'total_deaths', ascending = True)
df4
```

```python
df4.head(20).plot(kind = 'bar', color='purple')
plt.xlabel('Location', size = 5)
plt.ylabel('Total Death', size = 5)
plt.title("Continent with the highest Death")
plt.show()
```

```python
df5 = df.groupby(['location','population']) ['total_deaths'].sum().to_frame().sort_values('total_deaths', ascending = True)
df5
```

```python
df5.head(20).plot(kind = 'line', color='brown')
plt.xlabel('Location', size = 5)
plt.ylabel('Total Death', size = 5)
plt.title("Location with the highest Death")
plt.xticks(rotation=90)
plt.show()
```
