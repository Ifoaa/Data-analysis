```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('adult.csv')
df
```

```python
df.info()
```

```python
df = df.apply(lambda x: x.replace('?', 'None'))
df
```

```python
df.info()
```

```python
df.drop_duplicates(keep = False, inplace = True)
df
```

```python
df.info()
```

```python
df['Hours per week'] = df['hours.per.week']
df['Native country'] = df['native.country']
df
```

```python
df['capital gain'] = df['capital.gain']
df['capital loss'] = df['capital.loss']
df
```

```python
df1 = df.groupby(['sex', 'income']) ['hours.per.week'].count().to_frame().sort_values(by = 'sex', ascending = True)
df1
```

```python
df1.plot(kind = 'bar' )
plt.xlabel('Sex', color = 'red', size = 15)
plt.ylabel('Hours per week', color = 'green', size = 15)
plt.title('Sex with the highest number worked per hours per week')
plt.show()
```

```python
lesser_income = []

for i in df['income']:
    if i == '<=50K':
        lesser_income.append(i)
    else:
        lesser_income.append(None)

df['Lesser income'] = lesser_income
df
```

```python
higher_income = []
for i in df['income']:
    if i == '>50K':
        higher_income.append(i)
    else:
        higher_income.append(None)

df['Higher Income'] = higher_income
df
```

```python
# Get the Education degree with the highest capital gain while msking Lesser income

df2 = df.groupby(['education', 'Lesser income']) ['capital gain'].sum().to_frame().sort_values(by = 'capital gain', ascending = True)
df2
```

```python
df2.plot(kind = 'bar', color = 'green')
plt.xlabel('Education', size = 6)
plt.ylabel('Capital Gain', size = 6)
plt.title("Education degree making lesser income with the highest Capital Gain")
plt.plot()
```

```python
df3 = df.groupby(['education', 'Higher Income']) ['capital gain'].sum().to_frame().sort_values(by = 'capital gain', ascending = True)
df3
```

```python
df3.plot(kind = 'bar', color = 'red')
plt.xlabel('Education', size = 6)
plt.ylabel('Capital Gain', size = 6)
plt.title("Education degree making higher income with the highest Capital Gain")
plt.plot()
```

```python
df4 = df.groupby(['education', 'race']) ['Lesser income'].count().to_frame().sort_values(by = 'Lesser income', ascending = True)
df4.head(30)
```

```python
df4.head(20).plot(kind = 'bar', color = 'orange')
plt.xlabel('Education', size = 6)
plt.ylabel('Lesser income', size = 6)
plt.title("Race with the highest number earning lesser income")
plt.plot()
```
